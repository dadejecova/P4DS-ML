```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
```

```python
data = {'names':['steves','john','richard','sarah','randy','michael','julie'],
        'age':[20,22,20,21,25,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
df = DataFrame(data,columns=['names','age','gender','rank'])
df
```

```python
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Bar Chart')
plt.show()

#Seaborn
plot = sns.barplot(data=df, x='names', y='age')
plot.set_title("Bar Chart Seaborn")
plt.show()

```

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Line Chart')
plt.show()

#Seaborn
plot = sns.lineplot(data=df, x='names', y='age')
plot.set_title("Line Chart Seaborn")
plt.show()
```

```python
plt.pie(df['age'], labels=df['names'])
plt.title('Pie Chart')
plt.show()

colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'], labels = df['names'], colors = colors)
plt.title('Pie Chart Seaborn')
plt.show()
```

```python
# Libraries Import
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# Data
DF_obj = DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
# Data
DF_obj2 = DataFrame(np.arange(15).reshape(5,3))
DF_obj2
```

```python
# Concatenating Data
pd.concat([DF_obj, DF_obj2], axis=1)
```

```python
# Concatenating Data
pd.concat([DF_obj, DF_obj2])
```

```python
# Transforming Data
```

```python
# Dropping Data
DF_obj.drop([0,2])
```

```python
# Dropping Data
DF_obj.drop([0,2], axis=1)
```

```python
# Adding Data
series_obj = Series(np.arange(6))
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index = False)
added_datatable
```

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index = True)
added_datatable
```

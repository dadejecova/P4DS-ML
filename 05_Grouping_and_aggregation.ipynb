```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ruiromanini/mtcars")

print("Path to dataset files:", path)
```

```python
# it changes acording to the download path
cars = pd.read_csv('/root/.cache/kagglehub/datasets/ruiromanini/mtcars/versions/1/mtcars.csv')

#column names
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.head()
```

```python
# grouping by cylinders, 4 - 6 - 8
cars_groups = cars.groupby('cyl')
cars_groups.mean(numeric_only=True)
```

```python
# Treating missing values - working with misssing data in pandas
```

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
#Filling missing values using fillna(), replace() and interpolate()
```

```python
data = {'names':['steve', 'john', 'richard','sarah','randy','micheal','julie'],
        'age': [20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
# we need to create a data constructor using the data we create before

ranking_df = DataFrame(data)
ranking_df
```

```python
ranking_df.iloc[2:5,1] = np.nan
ranking_df.iloc[3:6,3] = np.nan
ranking_df.iloc[3,:] = np.nan

ranking_df
```

```python
# Pandas has several different funcions that are available to us for handling missing values.
# The fisr step for handling missing values is to detect if there are any missing values present in the dataset.
# In order to do that, Pandas provide us two functions, isnull and not null
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
# Filling the missing value with a single value
```

```python
ranking_df.fillna(0)
```

```python
# This is used to change missing values to the previous one
ranking_df.fillna(method='pad')
```

```python
ranking_df.fillna(method='bfill')
```

```python
ranking_df.interpolate(method='linear')
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all') 
```

```python
ranking_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```

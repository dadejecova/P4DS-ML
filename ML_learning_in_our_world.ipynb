```python
# Activities realized for an online course on LinkedIn Learning called "Machine Learning with Python: Foundations".
# How to Deal with Missing Data in Python

# During the process of data exploration,
# it's not uncommon to realize that the data we have is incomplete,
# missing data could arise as a result of changes in data collection methods,
# human error, bias, or simply lack of reliable input.
```

```python
import pandas as pd
students = pd.read_excel("/content/students.xlsx")
students
```

```python
mask = students['State'].isnull()
mask
```

```python
students[mask]
```

```python
# To Remove Missing Values
```

```python
students.dropna()
```

```python
students = students.dropna(subset = ['State', 'Zip'], how = 'all' )
students
```

```python
students.dropna(axis = 1)
```

```python
students = students.dropna(axis = 1, thresh = 10 )
students
```

```python
# To Resolve missing Data
```

```python
students = students.fillna({'Gender':'Female'})
students
```

```python
students = students.fillna({'Age':students['Age'].median()})
students
```

```python
mask = (students['City'] == 'Granger') & (students['State'] == 'IN')
mask
```

```python
students.loc[mask, :]
```

```python
students.loc[mask,'Zip'] = 46530
students
```

```python
mask = (students['City'] == 'Niles') & (students['State'] == 'MI')
students.loc[mask, 'Zip'] = 49120
students
```
